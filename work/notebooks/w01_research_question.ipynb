## Quick look at the data

The starter dataset contains **30,000 content pages**.

The correlation between **search volume** and **impressions** is approximately **0.001**, showing that search volume alone is not a reliable predictor of page impressions in this dataset.

The average CTR decreases substantially as search position becomes worse. For example, pages in the **page_1** position tier have an average CTR of about **0.355**, while pages in the **deep** position tier have an average CTR of about **0.055**.

These observations suggest that multiple observable signals should be considered together when prioritizing pages for review.

```python
# This cell is for CODE (numbers, a query, a check).
import os
import pandas as pd

if os.path.basename(os.getcwd()) == "notebooks":
    os.chdir("../..")

print("Working directory:", os.getcwd())

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Rows:", len(df))

corr = df["search_volume"].corr(df["impressions_90d"])
print("Search volume vs impressions correlation:", round(corr, 3))

ctr = (
    df[df["impressions_90d"] >= 100]
    .groupby("position_tier")["ctr"]
    .mean()
)

print("\nAverage CTR by position tier")
print(ctr.round(3))
```
