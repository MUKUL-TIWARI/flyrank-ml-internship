## Distribution Audit

The selected search metrics were reviewed to understand their distributions.

Search impressions and clicks typically show a right-skewed distribution where a small number of pages receive much higher traffic than the majority of pages.

This pattern suggests that summary statistics should be interpreted carefully and that tree-based models are appropriate for handling such distributions.

```python
import pandas as pd

distribution_summary = pd.DataFrame({
    "Metric": [
        "Impressions",
        "Clicks",
        "Average Position"
    ],
    "Observed Pattern": [
        "Right-skewed",
        "Right-skewed",
        "Approximately continuous"
    ]
})

distribution_summary
```
